### FIB

N,K -> begin with 1 pair; produce K additional pairs each N timesteps

```python
n,k = dataset = 36,2
```

```python
def fibby(n, k):
    if n==1:
        return 1
    if n==2:
        return k
    gen1 = fibby(n - 1, k)
    gen2 = fibby(n - 2, k)
    if n<=4:
        return (gen1+gen2)
    
    return (gen1+(gen2*k))

fibby(n,k)
```
